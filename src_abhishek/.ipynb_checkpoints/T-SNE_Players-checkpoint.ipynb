```python
from main_buckets import analyze_player
import pandas
```

```python
player_attributes = pandas.read_csv("..\data\Player_Attributes.csv")
playerid = player_attributes["player_api_id"].values.astype(int)
player_name_file = pandas.read_csv("..\data\Player.csv")
playernames = player_name_file["player_name"].values.astype(str)
playernamesids = player_name_file["player_api_id"].values.astype(int)
```

```python
playerid.shape
print(playernames)
print(playernamesids)
playerid[:5]
messi_idx = playernamesids==30981
ronaldo_idx = playernamesids==30893
buffon_idx = playernamesids == 30717
cassilas_idx = playernamesids == 30657
pique_idx = playernamesids == 37482
puyol_idx = playernamesids == 30661
print(ronaldo_idx)
```

```python
player_name_dict  = dict()
idx = 0
for id in playernamesids:
    player_name_dict[id] = playernames[idx]
    idx += 1
player_name_dict
```

```python
player_attributes_dict = []
player_names = []
for id in playerid:
    player_vector = analyze_player(id)
    player_attributes_dict.append(player_vector)
    player_names.append(player_name_dict[id])
    #break
```

```python
len(player_attributes_dict)
player_attributes_dict
```

```python
import numpy as np
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler


X = np.array(player_attributes_dict)
sc = StandardScaler()
X = sc.fit_transform(X)
print(X.shape)
model = TSNE(n_components=2, random_state=0)
```

```python
np.set_printoptions(suppress=True)
print X[messi_idx]
TX = model.fit_transform(X[:2500]) 
TX.shape
```

```python
from matplotlib import pyplot as plt

# plot the result
vis_x = TX[:, 0].ravel()
vis_y = TX[:, 1].ravel()

print vis_x.shape, vis_y.shape

plt.scatter(vis_x, vis_y)
#for label, x, y in zip(player_names[:2500], TX[:, 0], TX[:, 1]):
        #plt.annotate(label, xy=(x, y), xytext=(0, 0), textcoords='offset points')
print(X[messi_idx], X[messi_idx].shape)


model.fit(X[messi_idx].reshape(-1, 1))
T_messi = X[messi_idx]
print(T_messi)
plt.annotate("Messi", xy=(T_messi[:,0], T_messi[:,1]), xytext=(0, 0), textcoords='offset points')

model.fit(X[ronaldo_idx].reshape(-1, 1))
T_ronaldo = X[ronaldo_idx]
print(T_ronaldo)
plt.annotate("Ronaldo", xy=(T_ronaldo[:,0], T_ronaldo[:,1]), xytext=(0, 0), textcoords='offset points')

model.fit(X[buffon_idx].reshape(-1, 1))
T_buffon = X[buffon_idx]
print(T_buffon)
plt.annotate("Buffon", xy=(T_buffon[:,0], T_buffon[:,1]), xytext=(0, 0), textcoords='offset points')


plt.show()
```
